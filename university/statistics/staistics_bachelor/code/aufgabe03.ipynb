#### Aufgabe 3.1A

```python
liste = [3, 5, 12, 17, 23] #beispiel

#mittelwert berechnen
mw = sum(liste)/len(liste)
print(mw)

#zähler unter wurzel für varianzberechnung
zaehler=0
for i in liste:
   zaehler += (i-mw)**2
print(zaehler)

#varianz berechnen
varianz = (zaehler/len(liste))**0.5
print(varianz)

#standard-abweichung
sa = varianz**0.5

print(sa)
```

### Text einlesen und zuschneiden

```python
import os
import re
from tqdm import tqdm
```

```python
#pfad einfügen lassen und leere liste erstellen zum einlesen

#text = input("Bitte den genauen Pfad hier einfügen...")
text = r"C:\Users\Simon\Documents\Uni Würzburg\Digital Humanities\Bachelorarbeit\datasets\MeinK\MeinKampf_deutsch.txt"
```

```python
# Open the file in read mode
with open(text, encoding="utf-8") as f:
  # Read the entire content of the file
  text = f.read()
```

```python
#index textbeginn
matches = re.finditer(r"(?i)EINE ABRECHNUNG", text)

for match in matches:
  print(f"Index: {match.start()}")
```

```python
#index textende
matches = re.finditer(r"(?i)mit dem möglichen Erfolg verleiten sollte.", text)

for match in matches:
  print(f"Index: {match.start()}")
```

```python
text_cut = text[182252:1748264+len("mit dem möglichen Erfolg verleiten sollte.")]
print(text_cut)
```

------

### Bindestriche ersetzen

```python
text_cut = text_cut.rstrip()
```

```python
text_cut = text_cut.replace("- \n", "")
```

```python
text_cut = text_cut.replace("\n", "")
text_cut
```

```python
#gesamte textlänge
print(f'Anzahl der Wörter: {len(text_cut)}')
```

-------------

### Satzzeichen entfernen

```python
import nltk
nltk.download('wordnet') #is up-to-date!!!
```

```python
import string

punc_to_remove = string.punctuation  #Satzzeichen aus 'string.punctuation' verwenden

#Satzzeichen durch leere Zeichen zu ersetzen
table = str.maketrans('', '', punc_to_remove)

clean_text = text_cut.translate(table)

print(clean_text[:1000])
```

```python
#textlänge ohne satzzeichen
print(len(clean_text))
```

------

#### Erweitertes Text-Cleaning nötig

```python
from collections import Counter
```

```python
split_text = clean_text.split()
word_counts = Counter(split_text)

# Abrufen der 10 häufigsten Wörter
most_common_words = word_counts.most_common(100)

# Ausgabe der Ergebnisse
for word, count in most_common_words:
  print(f"{word}: {count}")
```

```python
#zeichen manuell ersetzen, das vom zeichenfiler zuvor nicht erkannt wurde
text = text.replace("„", "")

#einzelbuchstaben ersetzen (fehler von ocr)
text = re.sub(r"\s(a|b|c|d|e|f|g|h|i|j|k|l|m|n|o|p|q|r|s|t|u|v|w|x|y|z)\s", " ", text)

#zahlen ersetzen, die weniger als 4 stellen haben (4 stellen wegen jahreszahlen-berücksichtigung)
result = re.sub(r"\b\d{1,3}\b", " ", text)
```

#### Alle Texte verarbeiten

```python
import os
import glob
import re

input_year = int(input("Welches Jahr soll bearbeitet werden?"))

input_directory = "/home/simon/Dokumente/Uni Würzburg/Digital Humanities/Bachelorarbeit/datasets/my_trump/" + str(input_year) + "_election"
output_directory = "/home/simon/Dokumente/Uni Würzburg/Digital Humanities/Bachelorarbeit/datasets/my_trump/00_cleaned_files/" + str(input_year)

# Erstellen Sie das Ausgabeverzeichnis, falls es nicht existiert
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

pattern = os.path.join(input_directory, "*.txt")

# Frage einmal nach dem Zeitraum für alle Texte
while True:
    confirm = input_year
    if confirm <= 2022:
        break
    elif confirm > 2022:
        break
    else:
        print("Ungültige Eingabe. Bitte geben Sie 'ja' oder 'nein' ein.")

for path in glob.glob(pattern):
    # Öffnen Sie die Datei im Lesemodus
    with open(path, "r") as file:
        # Lesen Sie den Inhalt der Datei
        text_raw = file.read()

        # Bearbeiten Sie den Dateiinhalt nach Bedarf
        print(f'\n*****************************************************************************************************\n')
        print(f"Datei: {os.path.basename(path)}\n")
        print(f"Inhalt: {text_raw}")

        # Prüfen Sie die Bestätigung für den Zeitraum und wenden Sie entsprechende Muster an
        if confirm <= 2022:
            pattern = r'(?:President\s+Donald(?:\s+J\.)?\s+Trump?|Donald(?:\s+J\.)?\s+Trump?|President Trump):.*?\n(.*?)(?=\n(?:[^:]+:|$))'
            matches = re.findall(pattern, text_raw, re.MULTILINE | re.DOTALL) #vor 2022
            united = "\n\n".join(matches) #vor 2022
        elif confirm > 2022:
            pattern = r"(?:Former\s+)?(?:President\s+)?(?:Donald\s+)?Trump \(\d+:\d+\):\n(.*?)(?=\n\n(?:[^:\n]+ \(\d+:\d+\):|$))"
            matches = re.findall(pattern, text_raw, re.DOTALL)
            united = " ".join(matches)
        else:
            print("Ungültige Eingabe.")

        # Reguläre Ausdrücke zum Entfernen von Zeitangaben und eckigen Klammern
        text_cleaned = re.sub(r'\(\d+:\d+(:\d+)?\)', ' ', united) # Zeitangaben entfernen
        text_cleaned = re.sub(r'\[.*?\]', '', text_cleaned) # Eckige Klammern samt Inhalt entfernen

        # Speichern des bereinigten Textes in eine neue Datei im Ausgabeverzeichnis
        output_file = os.path.join(output_directory, f"{os.path.basename(path)}")
        with open(output_file, "w") as file:
            file.write(text_cleaned)

        print(f"Bereinigter Text wurde erfolgreich in die Datei {output_file} gespeichert.")

        print(text_cleaned)
        print(text_cleaned.split()[:10])
```

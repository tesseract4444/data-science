```python
import glob
import os
import re
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.text import Text
```

#### Alle Texte verarbeiten

```python
pattern = os.path.join("/home/simon/Dokumente/Uni Würzburg/Digital Humanities/Bachelorarbeit/datasets/my_biden/00_cleaned_files/2024", "*.txt")

#all_texts = []

for path in glob.glob(pattern):
    # Öffnen Sie die Datei im Lesemodus
    with open(path, "r") as file:
        # Lesen Sie den Inhalt der Datei
        text_raw = file.read()
        #all_texts.append(text_raw)

        # Bearbeiten Sie den Dateiinhalt nach Bedarf
        print(f'\n*****************************************************************************************************\n')
        print(f"Datei: {os.path.basename(path)}\n")
        print(f"Inhalt: {text_raw}")

        
        #regulärer Ausdruck zum Entfernen von Zeitangaben, Ziffernfolgen sowie der verschiedenen sprecher
        text = re.sub(r'\(\d+:\d+(:\d+)?\)', ' ', text_raw) #zeitangaben, die vom regex-ausdruck zuvor noch nicht gelöscht werden konnten löschen
        
        text = re.sub(r"\[.*?\]", '', text) #eckige klammern samt inhalt löschen
        #text = text.replace('\n', '') #entfällt, wegen oben
        
        print(text)
        print(text.split()[:10])


        #text speichern
        #filename = "EDIT_"+os.path.basename(path)
        filename = "all_files.txt"
        
        #ziel-ordner zum speichern
        file_path = "/home/simon/Dokumente/Uni Würzburg/Digital Humanities/Bachelorarbeit/datasets/my_biden/00_cleaned_files"
        
        #datei schreiben
        with open(os.path.join(file_path, filename), "a") as file:
            # Schreiben Sie den Text in die Datei
            file.write(text)
        
        print(f"gespeichert unter: {filename}")
        print(f'gespeichert im Ornder: {file_path}')
```

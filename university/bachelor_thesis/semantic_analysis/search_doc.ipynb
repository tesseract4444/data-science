#### Funktionen zur Durchführung der Analyse

```python
def read_directories(directories):
    documents = {}
    unique_documents = set()
    for directory in directories:
        for filename in os.listdir(directory):
            if filename.endswith(".txt"):
                filepath = os.path.join(directory, filename)
                with open(filepath, 'r', encoding='utf-8') as file:
                    documents[filename] = file.read()
    print(f'Gesamtzahl aller Dokumente: {len(documents)}')
    return documents

def find_term_sent(documents, search_term):
    results = []
    unique_docs = set()
    for filename, text in documents.items():
        sentences = sent_tokenize(text)
        for index, sentence in enumerate(sentences):
            if search_term in sentence:
                results.append((filename, index + 1))  #nummerierung beginnt bei 1#
                unique_docs.add(filename)
    print(f'\nDas Wort wird {len(results)} mal verwendet.')
    print(f'Das Wort taucht in {len(unique_docs)} einzelnen Dokumenten auf.')
    print('\nNachgewiesene Dokumente:')
    return results
```

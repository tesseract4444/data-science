#### 02 How SVMs work

- look for best hyperplane separation (which is the best line, the optimal line for separation)
  


- blue line looks overfit, green and orange one seem to be correct (but which one to choose?)



- find the widest possible (the possible largest) street between the datasets
- point(s) closest to the decision boundary define the support vectors



- SVM tries to make the margin as large as possible while also minimizing the amount of violations (points lie within the region defined by the support vectors)
- moving the margins further out means we have more points in our street (those are the margin violations)
- idea of SVM: decision boundary with support vectors running parallel to it while also trying to maximize the difference to the distance between the different datasets while also minimizing the amount of margin violations (number of points that go in between the red dash and green dash line our vectors)
- Hard Margins
    - keep data "off the streets"
    - sensitive to outliers
    - only works if data is fully separable (and linear)
    - out street now get much narrower as before (decision boundary get narrower)



- Soft Margins
    - data is allowed on the street
    - balance street width vs margin violation
    - likely generalizes better because it's less strict
    - we try to keep the street as wide as possible and we allow margin violations
    - here we are trying to keep the street as wide as possible (keeping distances between the support vectors as large as possible) while still minimizing margin violations
    - decision boundary in the image shows where we switch the decision boundaries, one of the yellow points lies on the opposite of the decision boundary (that point must be part of the purple group in this case)
    - danger of overfitting! -> but soft margins generalize better



- margin determined by C parameter (C is a regularization parameter)
    - the higher C -> the stricter margins
    - with C=1 we have less margin violations but the yellow point is still on the left (missclassified)
 


    - C=10 sets our point to the yellow points



- Hinge Loss (our loss function for the SVM)



- Prediction (pediction=0 means loss is 1; prediction=1 means loss is 0)



- Optimization
    - Zeta controls amount of margin violations (smaller -> less violations)
    - weight term controls size of the streets (smaller -> larger street)
    - => our goal is to minimize Zeta and the weights!
    - C controls trade-off between two objectives (decreasing C means the margin violations are less impactful and the weight parameters become more impactful) => balance is important

#### 03 When to use SVMs

- small to medium sized datasets
- hyperplane separable features (hyperplane that separates the features)
- sensitive to feature scales (changes the support vectors, feature scaling advisable like min-max/ordinal/standar scaler)
- C parameter with min-max-scaler as a transformation on our features

#### 03 Testing several

- What model will work best?
- process: train them, cross validation (using default parameters), pick the best performing models, do some grid search, narrow down the best model
- pick several models, compare baseline performance of each model
- pick 1-3 best performers, do finer tuning
- continue iteration on final model

#### 01 KNN

- neighborhood method
    - k-neighbors, vote for classification (prediction based on a vote for a distinct item e.g. "I'm a green instance")
    - ties are resolved by taking the instance that appears first (change the scoring)
- looser assumptions in KNN than with Log.Reg or SVM
- less work during training, more during prediction
- benefits a lot from dimensionality reduction





-----------------

#### 02 KNN Parameters

- k (n_neighbors)
- you don't want too many neighbors (makes it difficult to distinguish)
- weights (that each neighbor contributes) - votes from neighbors (e.g. weight their vote based on how close they are to my new instance, instances that are much closer to the new instance will carry more weight than the instances further away)
- distance metric
    - to determine k nearest neighbors
    - many different possibilities for metrics (flat space, Euclidean, Manhattan etc.)
    - Manhattan  (moving left/right then up/down)
      
    - Euclidean 
      
    - L_m
      

--------------------------

#### 03 KNN Parameters

- deep blue and yellow colors: training values
- light blue and pink colors: testing values







- increasing k makes the decision boundary smoother
- we can also change the distance metric which is called "p" (it also has an effect on our decision boundary)





- default: p=2 (Euclidean Distance)
- p=1: Manhattan Distance
- p > 2: Higher values of p give more weight to larger differences in features. So, points with significantly different values in even one feature will be considered further apart compared to when using lower p values.

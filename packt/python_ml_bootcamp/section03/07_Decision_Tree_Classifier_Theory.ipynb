- 
- leaf: final point in a tree after nothing has happened
- node: inbetween point
- True goes to the left, False goes to the right

#### 02 How does a decision tree learn?

- cost function (CART)
    - only 2 children per node (other algorithms can produce more)
    - formula: N as the number of samples in the original node divided by the number of samples on the left hand side multiplied by the gini score on the left hand side + SAME_THIN_ON_THE_RIGHT_HAND_SIDE
    - 
    - always a goal in machine learning: minimize the cost function
- greedy splitting (by sklearn)
    - the best split at each point is used, it doesn't go up and down nodes and reverse decisions (sometimes imperfect splits -> but good performance)
- how does it decide on the leaves / nodes
    - at each step takes subset of features to use test splits (e.g. 3/4 features)
    - best (slower but more accurate) vs random split (faster but more inaccurate)
    - best is more computaionally expensive, can also lead to overfitting (greedy)
    - random selects a random feature from the subset and finds best split (good against overfitting)
- goal: finding the best possible model that generalizes well on new data

#### 03 Decision tree parameters

- max features (number of features to consider for choosing to find test splits on)
    - sklearn randomly permutes features at each step too
    - random choices create different trees on the same dataset
    - best vs random
- parameters
    - maximum tree depth (how large our tree is, how far our tree goes) => a tree that is too big will overfit!
    - minimum leaf size (minimum number of samples we have in each leaf, e.g. instead of 40-80 we do a 50-70 leaf size, good against overfitting)
    - minimum samples required to split (if you don't want to end up like 10-10 split, so you define a 20 or else, good against overfitting)
    - maximum number of leaves (that our tree doesn't create a bunch of small leaves everywhere, so we have less decision paths)
    - minimum required impurity improvement to split (not making any unnecessary split where the impurity reduction is minimal, against overfitting)
    - ALL USED FOR REGULARIZATION, FOR GENERALIZATION, AGAINST OVERFITTING
    - => FINDING THE BEST OPTIMAL SOLUTION
- class probability
    - traverse tree breakdown proportion of samples in final leaf node (the green leaf in our example tree) that come from each class (e.g. bottom left proportions are 0,39/42, 3/42)

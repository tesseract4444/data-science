#### AdaBoost

- = adaptive boosting
- idea of boosting: series of weak learners
- series of weak learners become a strong learner
- series of models, focuses on learning the mistakes its predecessor makes (by focusing more on the samples they got wrong)
- starting with training data => True targets => feed both in our predictions => analyze what the model did wrong => updating the training weights => train a second model with the updated weights => do the same thing again and again ... => models will focus and learn of what its predecessors made
- 
- 
- sample weight updates
    - in sklearn weights are scaled using exponentials that depend on the size of the error as well as estimator weight
    - model 2 makes more things wright that model 1 made wrong (weight correction)
- estimator weights
    - r - estimator error
    - v - learning rate
    - estimator weight = learning rate * log * (1-estimator_error / estimator_error):
    - 
    - final prediction is weighted vote (depends on the estimator weights) or average (median in sklearn)
    - before: learning rate moves closer to the global minimum; here: learning rate scales the estimator weights (goal: reducing the learning rate to reduce the weight)

#### Regularization

- regularization
    - reduce the learning rate => include more estimators 
    - number of estimators by doing early stopping
    - choose a learning rate: about 0.1
    - base estimator (by adding regularization into the base estimator)
- performance
    - SAMME (extension to Ada that also allows multiclass) Discrete (for classification) vs Real (for predicting probability instead of a class)
    - Real uses predict_proba, which can  lead to better test set performance

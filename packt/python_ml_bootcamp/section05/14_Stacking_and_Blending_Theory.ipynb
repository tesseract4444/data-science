#### Data Splitting - Blending

- split training set into training and holdout set (because if we just use all of our training data at once, it's gonna be difficult to simultaneously also optimize our models aswell as optimize the aggregation model) => this is called "blending"
- layer 1 is trained on training set (models don't have to be of the same sort e.g. linear regression + ridge regression etc.)
- layer 2 is trained on holdout set
    - predictions are made on holdout set output from the trained models from layer 1
- 
- we'll then take our reduced training set and we're gonna train each model in this first layer on a reduced training set, training it to learn the true values.And then we're going to take the holdout set and for the holdout set, we're going to use that to train the second model
- the aggregation model is called a blender
- layer predictions are made sequentially (layer 1 -> layer 2, the outputs of layer 1 are fed as inputs to the second layer)
- we can still scale and train different models or just add new train models and then retrain our blender on the holdout set

----

#### Extra layers

- can add additional aggregation layers
- data now needs to split into 3 sets, instead of 2
- each output of each model in the first layer is fed into the input of every single model (so we have more models in layer 2)
- 3 layers here: base layer 1, aggregation layer 2, final aggregation layer 3
- propper training of all layers without thinking of informatino leakage or introducing other sorts of bias, noise or disturbances into our data/models
- the more layers you add the more you have to split your original training data
- higher number of layers require more data splits

#### Considerations

- stacking or blending can be a better approach than simple aggregations (than just performing a simple aggregation as mean or mode)

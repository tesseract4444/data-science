### Voting Ensembles



#### 01 Voting Classifiers

- series of predictions from different models that we've trained to solve the same problem
- several ML models, final prediction is
    - hard voting: mode of individual predictions (count the number of votes, each model cast their number of votes, you take the prediction that has the highest count)
    - soft voting: highest sum class probabilites across all classes (you output the probability belonging to a certain class => then aggregate these probabilities => you take the class that has the highest aggregate probability, advantage to hard voting: classifier has a less harder impact, models that are unsure about their prediction or that areless certain about which class it is will have less ofan impact on the final result than those models where thediscrepancy between the probabilities between the different classes is higher)

    - hard voting: mode of individual predictions

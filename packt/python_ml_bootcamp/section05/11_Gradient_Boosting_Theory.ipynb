#### Gradient Boosting

- series of weak learners become a strong learner
- series of model, focuses on learning the mistakes its predecessor makes (by learning the residual errors)
    - 
- Gradient Boosting takes the difference from our actual predictions to the true predictions and uses that as the targets for the second model that we are going to train, instead of the true targets it's trying to predict the residual errors that our first model has made, same for the third model etc.
    - 
- final prediction sums up all the predictions of each stage
- sum of predictions through all the models: [1.8 + 0.1 + 0.09, 3.2 + 0.82 + 0, 3.3 +-0.2 - 0.13]
    - 
    - final prediction: [1.99, 4.02, 2.97]
    - True values: [2, 4, 2]

-----------

#### Considerations

- how many models shoudl you add?
    - look for a minimum in validation error and perform early stopping (adding too many estimators can lead to overfitting)
- models trained in series, not as scalable bagging/pasting (we have to wait until the predecessor of the model is done with training)
- XGBoost (library)
    - optimized gradient boosting, makes training faster and whole approach more scalable

----------------

#### Regularization

- high number of estimators can lead to overfitting
    - best pracice to choose number of estimatos by early stopping
- shrinkage
    - reduces the impact that every model has on the final prediction (rather than summing up all the predictions, each model's impact/influence on th efinal prediction will be reduced by a certain factor)
    - best practice use a low learning rate (v <= 0.1) found to result in better test errors
    - lower v requires more estimators
    - image shows predictions ŷ, formula shows v that is lower than 0.1
    - 
- bagging
    - performing bagging + shrinkage can result in even better performance
    - bagging without shrinkage for Gradient Boosting is not as good
    - good bagging subsample size is around 0.5 (50%)
    - subspaces not as effective on performance, good for training/run time (reduces the number of features)

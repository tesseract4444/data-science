#### Bagging & Pasting


- remember: with ensembles we want diverse classifiers to have no correlation between errors
    1) use different ML models (different approaches to the model which can lead to a better performance)
    2) train the same model on different data (e.g. train on a reduced dataset instead of a full dataset)
    - 


------------------

- bagging (boostrap aggregation)
    - bootstrapping: we sample our samples, we have our original data and each model gets a subsample of our original data (each model gets different training data), random sampling with replacement
- pasting
    - it's just the same as bagging but this time: random sampling without replacement
- in both cases: our models get different training data and leran slightly different things, will give us a higher bias and a lower variance, when doing aggregation the results will reduce our bias without increasing the variance
- sample size of each training subset usually smaller than original training set (important for pasting, not as important for bagging)
- each model has higher bias, lower variance, but aggregation of model predictions counteract that
- voting classifier and voting regressor can scale easily
- model (training) is independent -> scalable through distribution (we can perform predictions across different machines)



------------

- can also randomly sample features, in addition to/instead of randomly observations (for high dimensional datasets)
    - bagging/pasting - randomly sampling observations
    - random subspaces - randomly sampling just features
    - random patches - randomly sampling both features observations
- good for high dimensional data (lots of features)

---------------

- out-of-bag evaluation (leftover samples used to test for the generalization error)
    - out-of-bag also used for Random Forest (using bagging procedure and then using voting classifier or voting regressor depending on which random forest implementation we use)
    - scoring each model against the samples that were not included in its training set
    - allows us to estimate generalization performance without additional validation set
    - not as good as cross-validation though (because with CV we are training all or evaluating all models on the same data and can perform aggregation afterwards)

#### Extra Trees

- extremely randomized trees
    - even more random than the random random forests
    - when training an ensemble of decision trees, also randomize the thresholds (splitter="random"), in addition to the training samples and feature subset to consider at each node
    - rather than looking for the best split at every single node, we just take a threshold split
    - more bias, lower variance, more diverse models

#### 02 What is regression?

- predicting numerical values from a continuous spectrum, or large discrete, number range
  
- applications like
    - sales
    - temperature
    - demands (e.g. for items in a store)
    - energy requirements
    - ...

#### 03 How does Linear Regression look like?

- estimate parameters of linear regression
- bias term theta (weights)                                              
  
- linear regression is just drawing the best straight line through the data
- best line is different for every class
  

----------------------

#### 04 How does Linear Regression learn?

- process: we start with our prediction then we see how wrong our prediction is then we move our weights so that we go into the right direction, we can do this by using our loss function or cost function to improve our predictions
- loss function
    - MSE (difference between true value and predicted value for every single observation and divide that by the number of observations, local minimum = global minimum which is good according to the data distribution)                                                          
      
      
    - MAE (absolute difference between prediction and true value, MAE doesn't punish the outliers as heavily as the MSE, local minimum = global minimum)                                      
      
      
    - y - target
    - ŷ - prediction
    - n - observations                                                 

-------------------------

####  05 Linear Regression Modifications

- a cost function is a loss function + an extra component
  
- Extra = L2 regularization -> Ridge Regression
- Extra = L1 regularization -> Lasso Regression
- Extra = L1+L2 regularization -> ElasticNet

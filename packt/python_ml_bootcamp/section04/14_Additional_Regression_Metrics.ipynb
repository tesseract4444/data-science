### Additional Regression Metrics

#### metrics
    

-------------

#### R2 score 
- 1 -> -inf
- 1 - fraction of the difference of each prediction to its true value then squared divided by the difference betweeen the true value and the average value
     

------------------

#### mean squared error 

- 0 -> inf (MSE used a lot for evaluation or training models, 0 is best because it's close to the true value, inf is bad because it's infinitely far away from the true value)
- MSE is the sum over the square of the difference between the true value to the predicted value)
  

-------------

#### mean absolute error

- 0 -> inf
- similar to the MSE but now absolut instead of squared
- 

------------

#### explained variance score

- 1 -> -inf
- similar to the R2 score
- the top one between the true value to the prediction,we look at the difference for eachobservation between the true value to the predictionminus the averageor the mean difference between each true value and each prediction

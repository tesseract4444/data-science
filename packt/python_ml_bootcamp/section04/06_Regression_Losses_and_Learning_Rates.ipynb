#### 02 Learning Rates

- constant: at some point there's no more learning
  
  

</br>

- adaptive (learning rate): when loss hasn't changed for at last i iterations => update learning rate, it starts at constant then our learning rate cuts by some factor (when it doesn't find anymore improvements), reducing the learning rate towards a minimum
  
  

</br>

- power (relation): function over number of iterations, scaling down a constant, power variable to choose the power relationship
- formula: t as the number of iterations, example (t or t0 could be 1,2,3,4, learning rate will scale the number of iterations to the power of t0)
  
- learning rate drops off very quickly towards the beginning and then flattens out over time
  

</br>

- Inverse: has a specific case of power, power is going to be to the power of 1, pure inverse relationship, multiplicative factor at the bottom
  
- graph: inverse relationship has a less extreme drop of the power 1, but it will still be more sudden in some cases (especially the adaptive one, depends on your model and how it's learning in different learning phases)
  

</br>

#### Different learning rates may work well for different sorts of problems depending on how quickly we actually reach our minimum region. Goal: Using a learning rate, that is not too slow and not too fast and will give us the optimal solution.

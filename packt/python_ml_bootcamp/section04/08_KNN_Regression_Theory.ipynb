#### KNN

- neighborhood method
    - k-neighbors, average for regression
- result is (weighted) average of target value of neighbors
- rather than using weights and votes to classifiy classes instead the KNN regressor uses surrounding neighbors to set an average value basically for the target we are looking for
- for example when we are looking for 3 or 4 neighbors it will use the target values of those neighbors and perform a weighted average (weighted, distance weight etc.)

- graph: LSTAT (x axis) and RM (y axis), color coded median house values, on the right hand side we start at lower and as we go top the pricest are the highest

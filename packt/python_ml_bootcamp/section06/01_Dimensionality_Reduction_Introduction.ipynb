#### Idea

- high dimensional (data sets have to learn a lot of features, overfit/underfit) data sets require a lot of training data, and can lead to suboptimal fits
- data often has underlying structure (we have to pick out the key components), and is not randomly distributed across all dimensions

---------

#### Curse of Dimensionality

- issues when dealing with high dimensional datasets
- training instances required to properly fit additional features scale exponentially with number of features
- feature space is very sparse, and new instances can be unlike previous instances
- high dimensionality: through a lot of feature engineering or creating polynomial features

-------------

#### Reducing Dimensionality

- projecting
    - project higher dimensional datasets onto a lower dimension
    - x, y -> z1 (2d -> 1d)
    - z1 is a coordinate axis in the projected plane
    - image: projecting a 2d dataset (left hand side) to a 1d dataset (right hand side)
    - 
    - x, y, z -> z1, z2 (3d -> 2d)
    - z1, z2 are coordinate axes in the projected plane
    - image: data is densely on a plane, orange and blue line (left image) as the new coordinate axes for the right image
    - 
    - x, y, z -> z1
    - z1 is a coordinate axis in the projected plane
    - image on the left with line to project the right hand side
    - so we go from 3 dimensions (= 3 features) to 1 dimension (= 1 feature)
    - 
- manifold learning
    - example: s curve, a 2d sheet curving inside a 3d surface
    - a manifold embedded into a higher dimension
    - manifold hypothesis: real world data usually lives in a lower dimensional manifold that is embedded into the higher dimensional (feature) space
    - we can try to learn the manifold and reduce onto coordinates that define the manifold
    - only useful if data behavior in manifold is simpler than in original dimensions - try through experimenting
    - easier to analyze data in relation to their coordinates to this manifold
    - 
    - in this following case original dimension is simpler (values are defined by z height), we can make better distinctions because we have those variations in the coloring
    - 
    - manifold learing is only useful if the task is simpler in the lower dimensions

#### Why should we think about Dimensionality Reduction?

- can reduce computation times
- **can** result in better performing models due to less sparsity + possibly filtering out noise

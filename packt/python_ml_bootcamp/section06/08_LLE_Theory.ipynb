#### Locally Linear Embedding

- find manifold by maintaining local distances between points (within the local neighborhood)
- isomap we cared about all geometric distances between all points, here we just focus on local neighborhood
- maintain global non-linear structure by just focusing on local linear distances between points (assuming distances on small scales can be approximated to be linear, even if manifold over large scales are non-linear)

--------

- find nearest neighbors for our data points
- create a matrix that tracks local distances (linear) between data points in the neighborhood, this allows reconstruction of a datapoint purely from its neighbors

#### Hessian LLE

- no more dealing with Euclidean distances
- local weight vector found through Hessian matrix 
- measures local curvature
- eigenvalue decomposition of squared Hessian taking d smallest eigenvalues (as a constant) for projection
- although computationally expensive, sklearn optimizations make it about as fast as other LLEs when output dimensions are small

------------

#### Considerations

- enough data that our manifold is properly sampled in all neighborhoods and locally linear distance approximations hold (however computation time scales with number of data points too)
- another cool thing: distance weight matrices are invariant to scaling, translation, or rotation of the point and those in its local neighborhood, meaning we get intrinsic geometric distances
- Modified LLE and Hessian LLE do not perform as well as data sizes scale up (take longer to train and fit)
- we need enough data points that our local linear assumptions are valid enough => not a good solution for small datasets

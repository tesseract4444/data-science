#### Non-Negative Matrix Factorization

- assumption
    - values in dataset are non-negative (>=0)
- approach
    - decompose dataset into two matrices that, when multiplied (WxH), approximate our original dataset
- m*n/k*n/m*k: rows * columns
- W is our reduced matrix (our transformed dataset)
- H are our components for transformation
- X is our feature trainig set



------------

#### Optimization

- minimize distances between original and deconstruced matrix
- dstance metric/loss function is SE for matrices
    - Frobenius norm (see formula)
- formula: squared error for each component, iterating over every element in our matrix and looking at how we can reconstruct it, measuring the squared distance from the reconstructed value to the actual true value



---------------

#### Regularization

- L1
    - Frobenius
    - lambda: regularization parameter for the regularization strength
    - summing up the absolute values for both the W matrix and for the H matrix
    - then taking the square
    - 
- L2
    - same as for the L1 but lambda now multiplied by 1/2
    - 
- Elastic Net
    - the extra term r is just our L1 ratio
    - a combination of L1 and L2
    - final cost function is our original loss from the Frobenius norm plus the l1 regularization components times the l1 ratio + the l2 regularization component times one minus l1 ratio
- sklearn actually also allows selection if regularization should be applied to H, W, or both

- reducing dimensions
    - to reduce dimensions we just project our data onto the hyperplane defined by our principal components
 
<br/>

- the math behind it
    - our principle components are then the first d rows of V^H
    - P: projected down to d dimensions
    - X is centered around 0
    - eigenvalue coefficients give us degree of explained variance by each corresponding principal component
    - X^H: Hermitian of X matrix
    - V^H: Hermitian of V matrix
    - matrix X and matrix S
    - eigenvalues: inform us about the degree of variance, must be sorted in descending order
    - eigenvector: corresponds to eigenvalue
    - 
    - **reconstruction**: simply multiply our projected matrix by our original eigenvector matrix, take our projected dataset and multiply that by the first D components of our original Hermitian of the matrix and that will give us back our reconstructed dataset
    - **evaluating hyperparameters**:
        - reconstruction error = looking at the MSE of distance betwween original points vs reconstructed points (you calculate the MSE for the distance betweeen the location of the data points in our original dataset versus the location of the data points of our reconstructed dataset) => then you get the measure of the MSE which is a way to measure the performance of the hyperparameter of the PCA
        - 
        - grid search: with supervidsed learning model as second step and dimensionality reduction as first step (= first layer)

- Kernel PCA
    - project our data to a d dimensional space, where d >= N, with N being our current dimensions
    - in a higher dimensional space our data can almost always be linearly separated, and we can separate our data into clusters using some suitable hyperplane
    - 
    - the following hyperplane depends on the kernel we use, and using the same kernel trick we also had for SVMs, we can calculate the resulting projections of our data
    - this lets us define non-linear hyperplanes to project our data onto
    - dataset now looks perfectly separated in image 2
    - 
    - reconstruction:
        - unlike linear PCA, in kernel PCA we never compute the components, just the projections
        - for reconstruction we then have to perform extra fitting to find points in our original space that lie close to our project points in higher dimensional space (inverse transform method)

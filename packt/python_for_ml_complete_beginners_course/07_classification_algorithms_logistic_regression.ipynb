### 01 Logistic Regression
- is used to predict the likelihood of a specific class or occurence
- conditions to use the LR: if the information is binary, if you require probalistic outcomes, when a linear decision boundary is required
- applications: forecast survival in wounded patients in the medical profession, estimate the likelihood of a person suffering a heart attack, predicting the likelihood of a procedure or product failing, predicting a homeowner's possibility of defaulting on a lender

### 03 Implementation in Python: Importing Libraries and Datasets

```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix
```

```python
#importing the dataset
dataset = pd.read_csv("user+data.csv")
X = dataset.iloc[:, [2,4]].values
y = dataset.iloc[:,4].values
```

```python
X
```

### 04 Implemetation in Python: Splitting Data into Train and Test Sets

```python
#training and testing data (divide the data into two parts)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=0)
```

### 05 Implementation in Python: Pre-Processing

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)
```

```python
X
```

```python
X_train
```

### 06 Implementation in Python: Training the Model

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state=0)
classifier.fit(X_train, y_train)
```

### 07 Implementation in Python: Results Prediction and Confusion Matrix

```python
y_pred = classifier.predict(X_test)
print(y_pred)
```

```python
#making the confusion matrix
import seaborn as sns

cm = confusion_matrix(y_pred, y_test)
print(cm)
#result: 68+32 correct predictions, 0 incorrect predictions
```

```python
sns.heatmap(cm)
```

### 08 Logistic Regression Versus Linear Regression

**Logistic Regression**
- we estimate the values of categorical variables using logistic regression
- we discover the S-curve (we use sigmoid function to discover the S-curve) in Logistic Regression and use it to categorize samples
- it is not necessary to have a linear connection between the dependent and independent variables in logistic regression
- there should be no collinearity between the indpendent variables in logistic regression
- can also be used for classification

**Linear Regression**
- we estimate the outcome of continuous variables using linear regression
- in linear regression, we look for the best fit line, which allows us to predict the outcome with ease
- it is necessary for the connection between the dependent and independent variables to be linear
- there is a possibility of collinearity between the independent factors in linear regression

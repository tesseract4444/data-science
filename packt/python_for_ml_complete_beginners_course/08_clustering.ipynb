### 03 K-Means Clustering Algorithm

K-Means Clustering
- it is an iterative algorithm that divides the unlabeled dataset into k different clusters in such a way that each dataset belongs only one group that has similar properties

Algorithms of K-Clustering
</br>Following is an algorithm of K-Means Clustering:
- partition clustering
- without any cluster-internal structre, K-Means divides data into non-overlapping groups
- within a cluster, examples are very similar
- various clusters have quite different examples

### 04 Elbow Method
Determination of K in K-means clustering
- The performance of the K-means clustering algorithm depends upon highly efficient clusters that it forms. But choosing the optimal number of clusters is a big task

Elbow method
- This method uses the concept of WCSSS value WCSS stands for Within Cluster Sum of Squares, which defines the total variations within a cluster. The formula to calculate the value of WCSS.
- formula: (distances between each data point * its centroid)Â² all summed up (C1, C2, C3 etc.) in different clusters

### 05 Steps of the Elbow Method
- it executes the K-means clustering on a given dataset for different K values (ranges from 1-10)
- for each value of K, calculates the WCSS value
- plots a curve between calculated WCSS values and the number of clusters K
- the sharp point of bend or a point of the plot looks like an arm, then that point is considered as the best value of K


### 06 Implementation in Python

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

plt.rcParams["figure.figsize"] = (16,9)

#creating a sample dataset with 4 samples
X, y = make_blobs(n_samples=800, n_features=3, centers=4)
```

```python
X
```

```python
#finding the optimal number of clusters using the elbow method
from sklearn.cluster import KMeans

wcss_list = [] #initializing the list for the values of WCSS

#10 iterations
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=42)
    kmeans.fit(X)
    wcss_list.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss_list)
plt.title("the elbow method")
plt.xlabel("number of clusters")
plt.ylabel("wcss_list")
plt.show()
```

### 07 Hierarchical Clustering

**What is Hierarchical Clustering?**
- it is an algorithm that groups similar objects into groups called clusters
- the endpoint is a set of clusters, where each cluster is distinct from another cluster and the objects within each cluster are broadly like each other

**Type of Hiearchical Clustering**
</br> There are two types of Hierarchical Clustering:
- divisive clustering: is a top to down approach in which observations start from large clusters and splits into smaller ones
- agglomerative clustering: is a bottom to top approach in which observations start from many clusters and are merged as one cluster in the end

### 08 Density-Based Clustering

**What is Density-based Clustering?**
- locates regions of high density and separates outliers
- the regions of high density get separated from the regions of low density

**What is DBScan?**
- the most powerful attibute of the DBScan algorithm is that it can find out any arbitrary shape cluster without getting effected by noise

**Advantages**
- it can discover arbitrary shaped cluster
- find cluster surrounded by different cluster
- robust towards noise detection

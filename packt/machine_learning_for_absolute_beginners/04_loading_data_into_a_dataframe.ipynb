#### Using Summary Statistics

```python
df = pd.read_csv("titanic.csv", header=0) #header=None removes the names of the columns
```

```python
df["Age"].min() #display minimum age
```

```python
df["Age"].mean() #mean (mittelwert)
```

```python
df[["Age", "Fare"]].mean() #mean age and fare (fare=ticketpreis)
```

```python
df["Survived"].sum() #sum of all survived passengers
```

```python
df["Fare"].sum() #sum of all the fares
```

```python
df["Fare"].max() #max of all the fares
```

```python
df_select = df[["PassengerId", "Survived", "Pclass", "Age","SibSp", "Parch", "Fare"]]
#accessing multiple data
```

```python
df_select.quantile(0.8) #maxima of 80 percent of all passengers
```

```python
df_select.max() #maxima of all passengers
```

```python
df.describe() #overview of all important statistics
```

```python
df["Age"].describe() #only statistics about age
```

#### The Concept of Method Chaining

```python
df.isnull() #returns a dataframe displaying missing values as a boolean
```

```python
df_temp = df.isnull() #shows amount of 0 values
```

```python
df_temp.sum()
```

```python
df.isnull().sum() #alternative
```

```python
type(df.isnull().sum())
```

```python
df.isnull().sum().sum() #sum of all
```

```python
df.dropna() #removes all 0 values
```

```python
df.dropna()["Age"].mean()
```

```python
df["Age"].mean() #original data (still containing 0 values)
```

```python
df.sort_values("Age", ascending=False).head(10) #sort all values by age
```

```python
df.sort_values("Age", ascending=False)[["Name",  "Age"]].head(10) #sort name values by age
```

```python
df.sort_values(["Sex", "Age"], ascending=[True, False])[["Sex", "Name", "Age"]]
#rows organized by the sex column (starting with all females) that are also sorted by age
```

```python
df.dropna().sort_values(["Sex", "Age"], ascending=[True, False])[["Sex", "Name", "Age"]]
```

```python
df.nlargest(10, "Age")[["Name", "Age"]]
#top 10 values sorted by age, only including name
```

```python
df.nsmallest(10, "Age")[["Name", "Age"]] #same thing as before for the samllest values
```

#### Filtering

```python
df["Sex"]=="female" #boolean for each row
```

```python
df[df["Sex"]=="female"] #filter values for females only
```

```python
df[df["Age"]>60]
```

```python
(df["Age"]>60) & (df["Sex"]=="female")
```

```python
df[(df["Age"]>60) & (df["Sex"]=="female")]
```

```python
df[(df["Age"]>60) | (df["Age"]<=20)]
```

```python
condition = (df["Age"]>60) | (df["Age"]<=20)
```

```python
df.loc[condition, ["Name", "Sex", "Age"]] #condition and relevant columns
```

#### Grouping

```python
df[["Sex", "Fare"]].groupby("Sex").mean()
```

```python
df[["PassengerId", "Pclass", "Age", "SibSp", "Parch", "Fare", "Survived"]].groupby("Survived").mean()
#groups survived in shown categories
```

```python
df.groupby("Embarked")["Survived"].count()
#groups embarked and counts the number of the survived
```

```python
df[["PassengerId", "Pclass", "Age", "SibSp", "Parch", "Fare", "Survived", "Sex"]].groupby("Sex").mean()
```

```python
df.groupby("Embarked")["Age"].agg(["count", "mean", "min", "max"])
#only examines special aggregaton functions (here: count, mean, min, max)
```

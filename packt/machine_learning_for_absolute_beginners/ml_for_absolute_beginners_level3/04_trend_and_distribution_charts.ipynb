## 02 Line Chart
- displays information as a series of data points
- demonstrates changes in value and observing te slopes of the lines moving up or down

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 2, 100) #creating 100 numbers between 0 and 2
fig, ax = plt.subplots()
ax.plot(x, x, label='linear')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("My Line Chart")
ax.legend()
plt.show()
```

```python
x = np.linspace(0, 2, 20) #creating 100 numbers between 0 and 2
fig, ax = plt.subplots()
ax.plot(x, x, label='linear', color='green', linestyle=':', linewidth=1, marker="^")
ax.plot(x, x**2, label='quadratic', color='blue', linestyle=':', linewidth=1, marker="x")
ax.plot(x, x**3, label='cubic', color='red', linestyle=':', linewidth=1, marker="o")
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("Simple Line Plot")
ax.legend()
plt.show()
```

```python
df = pd.DataFrame(dict(time=pd.date_range("2020-1-1", periods=100), value1=np.random.randn(100).cumsum()))
df['value2']=df['value1']+5
df.head()
```

```python
fig, ax = plt.subplots()
sns.lineplot(x="time", y="value1", data=df, ax=ax)
sns.lineplot(x="time", y="value2", data=df, ax=ax)
plt.show()
```

## 03 Area Chart
- similar to a line chart except that the area between the x axis and the line is filled with color

```python
x = np.linspace(0, 2, 20) #creating 100 numbers between 0 and 2
fig, ax = plt.subplots()
ax.plot(x, x, label='linear', color='green')
ax.plot(x, x**3, label='cubic', color='red')

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("Area Chart")
ax.legend()
plt.show()
```

```python
x = np.linspace(0, 2, 20) #creating 100 numbers between 0 and 2
fig, ax = plt.subplots()
ax.plot(x, x, label='linear', color='green')
ax.plot(x, x**3, label='cubic', color='red')

ax.fill_between(x, x**3, 0, color='red', alpha=0.4) #fill the area
ax.fill_between(x, x, 0, color='green', alpha=0.4) #fill the area
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("Area Chart")
ax.legend()
plt.show()
```

```python
x = np.linspace(0, 2, 20) #creating 100 numbers between 0 and 2
fig, ax = plt.subplots()
ax.plot(x, x, label='linear', color='green')
ax.plot(x, x**3, label='cubic', color='red')

ax.fill_between(x, x**3, x, color='yellow', alpha=0.4) #only fill area between the 2 axis

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("Area Chart")
ax.legend()
plt.show()
```

## 04 Stacked Area Chart
- a few charts that are vertically on top of one another rather than overlapping with one another (overlapping: e.g. area chart)

```python
x = np.linspace(0, 2, 20) #creating 100 numbers between 0 and 2
line1=x
line2=np.linspace(0.4, 0.4, 20)
line3=x
```

```python
x = np.linspace(0, 2, 20) #creating 100 numbers between 0 and 2
fig, ax = plt.subplots()
labels = ["linear1", "linear2", "linear3"]
ax.stackplot(x, line1, line2, line3, labels=labels)

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("Area Stacked Chart")
ax.legend(loc = "upper left")
plt.show()
```

```python
x = np.linspace(0, 2, 20) #creating 100 numbers between 0 and 2
fig, ax = plt.subplots()
labels = ["cubic", "linear2", "linear3"]
ax.stackplot(x, x**3, x, x, labels=labels)

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("Area Stacked Chart")
ax.legend(loc = "upper left")
plt.show()
```

## 05 Histogram Chart
- distribution analysis
- groups numerical data into segmented columns that are called bins
- the bins represent ranges of data
- the size of a bin will be proportional to the frequency of data points in that bin
- represent continuous data on the horizontal line

```python
titanic = sns.load_dataset("titanic")
titanic.head()
titanic.dropna(subset=["age"], inplace=True)
```

```python
fig, ax = plt.subplots()
ax.hist(titanic["age"], bins=5, edgecolor="black", color="red")
ax.set_xlabel("Ages")
ax.set_ylabel("People")
ax.set_title("Histogram Chart")
plt.show()
```

## 06 Density Curve Chart
- smoother view of the actual shape of the distribution
- visualize the underlying probability distribution of the data by drawing an appropriate continuous curve
- Kernel Density Estimation (KDE)

```python
fig, ax, = plt.subplots()
sns.histplot(titanic["age"], kde=True) #distplot got depricated, use histplot instead
ax.set_ylabel("Density")
ax.set_title("Density Chart")
plt.show()
```

```python
#fig, ax, = plt.subplots() #not necessary (returns empty plot)
sns.displot(titanic["age"], kind="kde") #distplot got depricated, use histplot instead
ax.set_ylabel("Density")
ax.set_title("Density Chart")
plt.show()
```

```python
#now let's see a kdeplot
fig, ax = plt.subplots()
sns.kdeplot(titanic["age"], fill=True, color="red", ax=ax) #distplot got depricated, use histplot instead
ax.set_ylabel("Density")
ax.set_title("Density Chart")
plt.show()
```

```python
#now let's regard a kdeplot
fig, ax = plt.subplots()
male_group = titanic[titanic["sex"]=="male"]
female_group = titanic[titanic["sex"]=="female"]

sns.kdeplot(titanic["age"], fill=True, color="blue", ax=ax)
sns.kdeplot(female_group["age"], fill=True, color="red", ax=ax)

ax.set_ylabel("Density")
ax.set_title("Density Chart")
ax.set_xbound(0) #prevent negative values (start x axis from 0)
plt.show()
```

##  07 Box-and-Whisker Chart
- compare several distributions against each other

```python
import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset("iris")
fig, ax = plt.subplots()
sns.boxplot(iris["sepal_length"], ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.boxplot(data=iris, ax=ax)
```

#### boxplot structure
- 50% of the data points are inside the box
- 25% data points above the box
- 25% data points below the box
- line in the middle of the box: median
- if the data distribution is not symmetric the median would not be in the middle
- small box: 50% of the values are around the same small range with less variation in the data set
- large box: a lot of variation within the data points
- vertical line (whisker): used to capture the range of the remaining 50% data, also showing the max and min value
- outliers: data points that are not fitting well to the distribution

#### violin chart
- combine a boxplot with a density plot

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.violinplot(data=iris, ax=ax)
plt.show()
```

##  08 Bee-swarm Chart
- alternative to the box chart
- showing the overall distribution of a variable and the individual data points
- makes it easier to spot outliers

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.swarmplot(x="day", y="total_bill", data=tips, ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips, ax=ax)
plt.show()
```

```python
#combining the violinplot with the swarmplot
fig, ax = plt.subplots(figsize=(8,8))
sns.violinplot(x="day", y="total_bill", data=tips, ax=ax)
sns.swarmplot(x="day", y="total_bill", data=tips, ax=ax)
plt.show()
```

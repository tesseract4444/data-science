## 01 Introduction
- correlation: how two or more variables are related
- correlation can be linear or non-linear, positive or negative, strong or weak, a combination of all (strong positive linear chart)
- correlation charts: Scatter, Correlogram, Heatmap, Hexbin map

## 02 Scatter Chart
- presenting relationships between two numeric variables by using simple dots
- collection of points on 2D space
- suggest a variety of correlation options between variables

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
fig, ax = plt.subplots(figsize=(7,7))
ax.scatter(x="total_bill", y="tip", data=tips) #scatter function for scatter chart
ax.set_xlabel("total_bill")
ax.set_ylabel("tip")
ax.set_title("Scatter Chart")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,7))
ax.scatter(x="total_bill", y="tip", color="green", marker="x", data=tips) #scatter function for scatter chart
ax.set_xlabel("total_bill")
ax.set_ylabel("tip")
ax.set_title("Scatter Chart")
plt.show()
```

```python
tips.head()
```

```python
colors = tips["sex"].map({"Male":"black", "Female":"red"}) #map writes all males as black and all females as red
colors.head(10)
```

```python
fig, ax = plt.subplots(figsize=(7,7))
ax.scatter(x="total_bill", y="tip", color=colors, data=tips) #scatter function for scatter chart
ax.set_xlabel("total_bill")
ax.set_ylabel("tip")
ax.set_title("Scatter Chart")
plt.show()
```

```python
colors=tips["size"].map({1:"black", 2:"pink", 3:"green", 4:"yellow", 5:"red", 6:"blue"})

fig, ax = plt.subplots(figsize=(7,7))
ax.scatter(x="total_bill", y="tip", color=colors, data=tips) #scatter function for scatter chart
ax.set_xlabel("total_bill")
ax.set_ylabel("tip")
ax.set_title("Scatter Chart")
plt.show()
```

```python
#display different sizes
colors=tips["size"].map({1:"black", 2:"pink", 3:"green", 4:"yellow", 5:"red", 6:"blue"})
sizes=tips["size"]*20

fig, ax = plt.subplots(figsize=(7,7))
ax.scatter(x="total_bill", y="tip", sizes=size, color=colors, data=tips) #scatter function for scatter chart
ax.set_xlabel("total_bill")
ax.set_ylabel("tip")
ax.set_title("Scatter Chart")
plt.show()
```

```python
#light red to strong red to distnguish between the tip sizes
fig, ax = plt.subplots(figsize=(7,7))
ax.scatter(x="total_bill", y="tip", c="tip", cmap="Reds", data=tips) #scatter function for scatter chart

ax.set_xlabel("total_bill")
ax.set_ylabel("tip")
ax.set_title("Scatter Chart")
plt.show()
```

```python
#heat measure for tips
fig, ax = plt.subplots(figsize=(7,7))
im = ax.scatter(x="total_bill", y="tip", c="tip", cmap="plasma", data=tips) #scatter function for scatter chart
fig.colorbar(im, ax=ax)

ax.set_xlabel("total_bill")
ax.set_ylabel("tip")
ax.set_title("Scatter Chart")
plt.show()
```

## 04 Heatmap
- showing relationships between two variables, one plotted on each axis, using colors
- observe how cell colors change across each axis to identify patterns
- the cells either contain color-coded categorical data or numerical data that is based on a color scale

```python
import numpy as np
import seaborn as sns

sns.set()
data = np.random.rand(4,4)
ax = sns.heatmap(data, cmap="YlGnBu")
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.corr() #show correlations between the data .corr()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(titanic.corr(), annot=True, fmt=".2f", ax=ax)
plt.show()
#result: which variables have stronger correlations between each other
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(iris.corr(), annot=True, fmt=".2f", ax=ax)
plt.show()
#result: e.g. petal_length highly correlates with petal_width (value: 0.96 which is close to 1.0)
```

```python
flights = sns.load_dataset("flights")
flights.head(10)
```

```python
len(flights["year"].unique())
```

```python
flights["year"].unique()
```

```python
#pivot: creates a new table with an invidually needed structure
new_flights = flights.pivot(index="month", columns="year", values="passengers")
new_flights.head()
```

```python
#color of each cell should be based on the number of passengers
fig, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(new_flights, cmap="magma", ax=ax)
plt.show()
#result: which variables have stronger correlations between each other
```

```python
#color of each cell should be based on the number of passengers
fig, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(new_flights, linewidths=.8, cmap="magma", ax=ax) #linewidths: splits cells by lines
plt.show()
#result: which variables have stronger correlations between each other
```

```python
#showing the exact numbers in each cell with annot=True and fmt="d"
fig, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(new_flights, annot=True, fmt="d", linewidths=.8, cmap="magma", ax=ax)
plt.show()
#result: which variables have stronger correlations between each other
```

## 06 Hexbin Map
- for special cases: huge datasets (data points that overlap each other)
- hexbin chart: puts points in logical hexagonal regions (color gradients showing the intensity of points in such a hexagonal area)

```python
n = 10000
df = pd.DataFrame({"x": np.random.randn(n),
                    "y": np.random.randn(n)})
df.head()
```

```python
fig, ax = plt.subplots(figsize=(7,7))
ax.scatter("x", "y", data=df)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,7))
ax.hexbin("x", "y", gridsize=40, cmap="inferno", data=df) #gridsize: 40 hexagonal regions
plt.show()
#result: density = number of points in such a hexagonal region
```

```python
fig, ax = plt.subplots(figsize=(10,7))
im = ax.hexbin("x", "y", gridsize=20, cmap="inferno", data=df) #gridsize: 40 hexagonal regions
fig.colorbar(im, ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,7))
im = ax.hexbin("x", "y", gridsize=20, cmap="inferno", edgecolors="grey", data=df) #edgecolors: add edges
fig.colorbar(im, ax=ax)
plt.show()
```

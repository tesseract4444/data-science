##  03 Bar Plot

####  Bar Plot - 1 Zeile

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://www.sololearn.com/uploads/ca-covid.csv")
df.drop('state', axis=1, inplace=True)
df['date'] = pd.to_datetime(df['date'], format="%d.%m.%y")
df['month'] = df['date'].dt.month
df.set_index('date', inplace=True)

(df.groupby('month')['cases'].sum()).plot(kind="bar") #'bar' zeichnet s채ulen
plt.savefig('plot.png')
```

####  Bar Plot - mehrere Zeilen

##  04 Box Plot & Histogramm

#### Box Plot
- wird benutzt, um das Ergebnis der describe()-Funktion zu visualisieren
- Beispiel: F채lle im Juni, gr체ne Linie zeigt Median, Box zeigt upper und lower Quartile (25%-Grenzen), Kreis zeigt Sonderf채lle an, schwarze Linie zeigt min/max an

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://www.sololearn.com/uploads/ca-covid.csv")
df.drop('state', axis=1, inplace=True)
df['date'] = pd.to_datetime(df['date'], format="%d.%m.%y")
df['month'] = df['date'].dt.month
df.set_index('date', inplace=True)

df[df["month"]==6]["cases"].plot(kind="box")
plt.savefig('plot.png')
```

#### Histogramm
- zeigt Verteilung der Daten an
- 채hnlich zur Bar(Balken)-Grafik
- zeigen eher H채ufigkeit einer Gruppe von Daten statt einziger Datenpunkt
- keine Freir채ume zwischen den Balken (alles ist verbunden)

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://www.sololearn.com/uploads/ca-covid.csv")
df.drop('state', axis=1, inplace=True)
df['date'] = pd.to_datetime(df['date'], format="%d.%m.%y")
df['month'] = df['date'].dt.month
df.set_index('date', inplace=True)

df[df["month"]==6]["cases"].plot(kind="hist")
plt.savefig('plot.png')
```

## 05 Area Plot, Scatter Plot, Pie Chart

#### Area Plot
- zeichnet ausgef체llte Bereiche unter einer Linie
- werden default gestackt, daher "stacked=False"

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://www.sololearn.com/uploads/ca-covid.csv")
df.drop('state', axis=1, inplace=True)
df['date'] = pd.to_datetime(df['date'], format="%d.%m.%y")
df['month'] = df['date'].dt.month
df.set_index('date', inplace=True)

df[df["month"]==6][["cases", "deaths"]].plot(kind="area", stacked=False)
plt.savefig('plot.png')
```

####  Scatter Plot
- zeigt Beziehung zwischen 2 Werten (Variablen) an
- Beispiel: wie F채lle und Tode zusammenh채ngen
- Datenpunkte streuen sich (scatter) innerhalb des Graphs

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://www.sololearn.com/uploads/ca-covid.csv")
df.drop('state', axis=1, inplace=True)
df['date'] = pd.to_datetime(df['date'], format="%d.%m.%y")
df['month'] = df['date'].dt.month
df.set_index('date', inplace=True)

df[df["month"]==6][["cases", "deaths"]].plot(kind="scatter", x='cases', y='deaths')
plt.savefig('plot.png')
```

#### Pie Chart
- stellen i.d.R. Prozente oder proportionale Zusammenh채nge dar
- werden normalerweise verwendet, wenn man bis zu 6 verschiedene Kategorien hat

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://www.sololearn.com/uploads/ca-covid.csv")
df.drop('state', axis=1, inplace=True)
df['date'] = pd.to_datetime(df['date'], format="%d.%m.%y")
df['month'] = (df['date'].dt.month_name()).str[:3]
df.set_index('date', inplace=True)

df.groupby('month')['cases'].sum().plot(kind="pie") #kine="pie" erstellt kreisdiagramm
plt.savefig('plot.png')
```

## 06 Plot Formatting

#### Plot Formatting Basics

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://www.sololearn.com/uploads/ca-covid.csv")

df.drop('state', axis=1, inplace=True)
df['date'] = pd.to_datetime(df['date'], format="%d.%m.%y")
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df.set_index('day', inplace=True)

df = df[df['month']==6]

df[['cases', 'deaths']].plot(kind="line", legend=True) #'legend' zeigt legende an
plt.xlabel('Days in June') #auch als ylabel setzbar: vertauscht dann achsen
plt.ylabel('Number') ##auch als xlabel setzbar: vertauscht dann achsen
plt.savefig('plot.png')
```

#### Die suptitle()-Funktion
- setzt einen Diagramm-Titel

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://www.sololearn.com/uploads/ca-covid.csv")

df.drop('state', axis=1, inplace=True)
df['date'] = pd.to_datetime(df['date'], format="%d.%m.%y")
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df.set_index('day', inplace=True)

df = df[df['month']==6]

df[['cases', 'deaths']].plot(kind="area", legend=True)
plt.xlabel('Days in June')
plt.ylabel('Number')
plt.suptitle("COVID-19 in June")
plt.savefig('plot.png')
```

#### Diagramm-Farben wechseln

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://www.sololearn.com/uploads/ca-covid.csv")

df.drop('state', axis=1, inplace=True)
df['date'] = pd.to_datetime(df['date'], format="%d.%m.%y")
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df.set_index('day', inplace=True)

df = df[df['month']==6]

#color wird festgelegt als liste, bestehend aus 2 hexcodes/farbcodes
df[['cases', 'deaths']].plot(kind="area", legend=True, stacked=False, color=['#1970E7', '#E73E19'])
plt.xlabel('Days in June')
plt.ylabel('Number')
plt.suptitle("COVID-19 in June")
plt.savefig('plot.png')
```

### Analyze columns: count, amount, percentage

#### functions

```python
#column_name = input()

def count_all(df, column_name):#count languages
    return df[column_name].value_counts()


def missing_values(df, column_name):
#count nan (empty) entries
    miss_vals = df[column_name].isna().sum()
    return f'{miss_vals} missing values'


def percentage_missing_values(df, column_name):
#calculate percentage of missing values
    miss_vals = df[column_name].isna().sum()
    amount_miss_vals = miss_vals/df.shape[0]
    return f'{round(amount_miss_vals*100, 2)}% missing values'


def sum_of_all(df, column_name):
    sum_all_lang = df[column_name].value_counts().sum()
    return f'{sum_all_lang} {column_name}(s)'
    

def percentage_sum_of_all(df, column_name):
    sum_all_lang = df[column_name].value_counts().sum()
    percent_all_lang = [i/sum_all_lang for i in df[column_name].value_counts().values]
    return f'percentage: {percent_all_lang}'

def plot_the_values(df, column_name):
    #plot the values
    plt.figure(figsize=(12, 6))
    plt.xticks(rotation=90)
    cmap = plt.cm.jet
    colors = cmap(np.linspace(1, 0, len(df[column_name].value_counts().values)))
    plt.bar(df[column_name].value_counts().index, df[column_name].value_counts().values, color=colors)
    plt.title(column_name)
    plt.xlabel(column_name)
    plt.ylabel("amount")
    plt.show()
    
    
def plot_the_percentage(df, column_name):
    sum_all_lang = df[column_name].value_counts().sum()
    percent_all_lang = [i/sum_all_lang for i in df[column_name].value_counts().values]
    plt.figure(figsize=(12, 6))
    plt.xticks(rotation=90)
    cmap = plt.cm.jet
    colors = cmap(np.linspace(1, 0, len(percent_all_lang)))
    plt.bar(df[column_name].value_counts().index, percent_all_lang, color=colors)
    plt.title(column_name)
    plt.xlabel(column_name)
    plt.ylabel("amount in %")
    ax = plt.gca()
    ax.set_ylim(0, 1)
    plt.show()
```

#### Languages

```python
print(count_all(df, "Language"))
print(missing_values(df, "Language"))
print(percentage_missing_values(df, "Language"))
print(sum_of_all(df, "Language"))
print(percentage_sum_of_all(df, "Language"))
print(plot_the_values(df, "Language"))
print(plot_the_percentage(df, "Language"))
```

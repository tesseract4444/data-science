```python
import pandas as pd
import random
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#generate random values for fixed costs, variable costs, and price per unit
fixed_costs = random.randint(3000, 10000)  #fixed costs between 3000 and 10000 €
variable_cost_per_unit = random.randint(10, 50)  #variable cost per unit between 10 and 50 €
price_per_unit = random.randint(30, 100)  #price per unit between 30 and 100 €

#formula for the break-even point
bep = fixed_costs / (price_per_unit - variable_cost_per_unit)
    
#maximum number of units for calculation
max_units = 500
    
#generate a list of units sold (from 0 to max_units)
units = list(range(0, max_units + 1))
    
#calculate the revenue for each number of units sold
revenue = [unit * price_per_unit for unit in units]
    
#calculate the total costs (fixed costs + variable costs per unit * units sold)
total_costs = [fixed_costs + variable_cost_per_unit * unit for unit in units]
    
#create a dataframe
df = pd.DataFrame({
    'Units Sold': units,
    'Revenue (€)': revenue,
    'Total Costs (€)': total_costs
    })
    
#print the random values and the dataframe
print(f"Randomly generated values:\nFixed Costs: {fixed_costs} € \nVariable Costs per Unit: {variable_cost_per_unit} € \nPrice per Unit: {price_per_unit} €\n")
print(df)

#print the break-even point
print(f"\nThe Break-Even Point is at {bep:.2f} units sold.")
```

```python
#plot the data using seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x='Units Sold', y='Revenue (€)', data=df, label='Revenue', color='blue', linewidth=2)
sns.lineplot(x='Units Sold', y='Total Costs (€)', data=df, label='Total Costs', color='red', linewidth=2)
    
#add a vertical line for the bep
plt.axvline(x=bep, color='green', linestyle='--', label=f'Break-Even Point at {bep:.2f} units')
    
#define plot
plt.title('Break-Even Analysis', fontsize=16)
plt.xlabel('Units Sold', fontsize=12)
plt.ylabel('Amount (€)', fontsize=12)
plt.legend()
    
#display the plot
plt.show()
```

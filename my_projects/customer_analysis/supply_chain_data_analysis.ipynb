```python
import pandas as pd
import numpy as np
from collections import Counter
import random
```

source of the "products.csv": https://gist.github.com/petroniocandido/dbab9321b7d5770b7549682436bb2073

```python
product_list = pd.read_csv("products.csv")
product_list = product_list["product_name"]
print(product_list)

random_products = random.choices(product_list, k=100)
```

```python
#random supply chain data
data = {
    "product": random_products,
    "stock": np.random.randint(5, 100, size=100),  # Stock levels between 5 and 100
    "delivery_time (days)": np.random.randint(1, 15, size=100),  # Delivery times between 1 and 15 days
    "sales_per_month": np.random.randint(1, 500, size=100)  # Sales between 1 and 500
}

df_supply = pd.DataFrame(data)
```

```python
#20 low-stock products (< 20 units)
low_stock = df_supply[df_supply["Stock"] < 20]

#average, min delivery times & max delivery times
avg_delivery_time = df_supply["delivery_time (days)"].mean()
max_delivery_time = df_supply["delivery_time (days)"].max()
min_delivery_time = df_supply["delivery_time (days)"].min()

#best- and worst-selling products
top_sellers = df_supply.nlargest(10, "sales_per_month")
low_sellers = df_supply.nsmallest(10, "sales_per_month")
```

```python
print(f"Average delivery time: {avg_delivery_time:.2f} days")
print(f"Longest delivery time: {max_delivery_time} days")
print(f"Shortest delivery time: {min_delivery_time} days")
```

```python
print("\nProducts with low stock:")
low_stock.sort_values("Stock", ascending=True)
```

```python
print("\nTop 5 best-selling products:")
top_sellers
```

```python
print("\nBottom 5 worst-selling products:")
low_sellers
```
